```python
import langchain
import faiss
import streamlit
from sentence_transformers import SentenceTransformer
from huggingface_hub import hf_hub_download

print("✅ LangChain:", langchain.__version__)
print("✅ FAISS: OK")
print("✅ All libraries working!")
```

```python
from langchain_community.document_loaders import PyPDFLoader

# Apna PDF path yahan daal
loader = PyPDFLoader("Vikas_Mali_Simple_Resume (1).pdf")
pages = loader.load()

print(f"✅ Total pages loaded: {len(pages)}")
print(f"📄 First page preview:\n{pages[0].page_content[:500]}")
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=500,
    chunk_overlap=50
)

chunks = text_splitter.split_documents(pages)

print(f"✅ Total chunks created: {len(chunks)}")
print(f"📄 First chunk:\n{chunks[0].page_content}")
```

```python
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS

# Embedding model load karo
embeddings = HuggingFaceEmbeddings(
    model_name="all-MiniLM-L6-v2"
)

# Vector store banao
vectorstore = FAISS.from_documents(chunks, embeddings)

print("✅ Vector store ready!")
print(f"✅ Total vectors stored: {vectorstore.index.ntotal}")
```

```python
from transformers import pipeline

# Fixed pipeline
qa_pipeline = pipeline(
    "text-generation",
    model="gpt2",
    max_new_tokens=100
)

# Retriever banao
retriever = vectorstore.as_retriever()

# Question poochho
question = "What is Vikas Mali CGPA?"
docs = retriever.invoke(question)

# Context lo
context = docs[0].page_content

# Answer lo
result = qa_pipeline(
    f"Context: {context}\nQuestion: {question}\nAnswer:"
)

print(f"❓ Question: {question}")
print(f"✅ Answer: {result[0]['generated_text']}")
```

```python
from langchain_groq import ChatGroq
from langchain_community.document_loaders import PyPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS

# Step 1 - PDF load
loader = PyPDFLoader("Vikas_Mali_Simple_Resume (1).pdf")
pages = loader.load()
print(f"✅ Pages loaded: {len(pages)}")

# Step 2 - Chunks
splitter = RecursiveCharacterTextSplitter(
    chunk_size=500,
    chunk_overlap=50
)
chunks = splitter.split_documents(pages)
print(f"✅ Chunks created: {len(chunks)}")

# Step 3 - Embeddings + Vector store
embeddings = HuggingFaceEmbeddings(
    model_name="all-MiniLM-L6-v2"
)
vectorstore = FAISS.from_documents(chunks, embeddings)
print("✅ Vector store ready!")

# Step 4 - LLM
llm = ChatGroq(
    api_key="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",
    model_name="llama-3.3-70b-versatile"
)
print("✅ LLM ready!")

# Step 5 - RAG Function
def ask_question(question):
    docs = vectorstore.as_retriever().invoke(question)
    context = "\n".join([doc.page_content for doc in docs])
    response = llm.invoke(f"""
Answer the question based on context only.
If answer not in context, say 'I dont know'.

Context: {context}
Question: {question}
Answer:
""")
    return response.content

# Step 6 - Questions poochho!
print("\n" + "="*50)
print(ask_question("What is Vikas Mali CGPA?"))
print("="*50)
print(ask_question("What projects has Vikas Mali done?"))
print("="*50)
print(ask_question("What are Vikas Mali skills?"))
```
